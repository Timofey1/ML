```python
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
import pandas as pd
df = pd.read_csv("brooklyn_sales_map.csv")
```

```python
df = df.drop(["Address"], axis=1)
```

```python
df = df.drop_duplicates("address")
```

```python
all_cols = df.columns
```

```python
plt.figure(figsize=(10,30))
missing = pd.DataFrame({'column':df.columns ,'na_percent':df.isnull().sum()/len(df)*100})
missing.sort_values('na_percent',inplace=True)
plt.barh(missing['column'],width=missing['na_percent']);
```

```python
cols_with_missing = [col for col in df.columns
                     if df[col].isnull().any()]
```

```python
drop=[]
for col in cols_with_missing:
    if df[col].isnull().value_counts().loc[True] > len(df)/2:
        drop.append(col)
```

```python
df = df.drop(drop, axis=1)
```

```python
cols_with_missing = [col for col in df.columns
                     if df[col].isnull().any()]
```

```python
numerical_miss_cols = [cname for cname in df[cols_with_missing] if df[cname].dtype in ['int64', 'float64']]
```

```python
s = (df[cols_with_missing].dtypes == 'object')
object_cols = list(s[s].index)
print(object_cols)
```

```python
for col in numerical_miss_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df.drop(["sale_price"], axis=1)
y = df.sale_price
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.7, test_size=0.3)
```

```python
model = RandomForestRegressor(random_state=42,n_estimators=100)
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, preds)
```
